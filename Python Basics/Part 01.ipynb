## Variables, Expressions, and Statements

### Constants 
Fixed values such as numbers, letters and strings. Their value does not change.;
Eg.: Numeric constants - 123,65;
     String constants  'Hello' "World";

```
print(123)
print("Hello World")
```

### Variables


A name place in the memory where a programmer can store data and later retrieve it using the variable name. The content of a variable can be changed later.;
Eg.: x = 12.2; y = 14;

```
x = 12.2
y = 14
print(x)
print(y)
```



*   Variable names can start with letter or underscore only.
*   Variable names can contain letters, numbers and underscores only.
*   Variable should be named sensibly where name means something. Avoid naming them as a,b,c etc.






```
hours = 35.0
rate = 12.50
pay = hours * rate
print(pay)
```

```
print(5/2)
print(5//2)
```
