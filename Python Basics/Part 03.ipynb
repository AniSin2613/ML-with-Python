```
# reading a file
# (3) readline() approach
flanders_file = open('/content/drive/My Drive/Python - Dimensionless/Python Basics/flanders.txt', 'r')
flanders_poem = ''

line = flanders_file.readline()
while line != '':
  flanders_poem = flanders_poem + line
  line = flanders_file.readline()
flanders_file.close()
print(flanders_poem)
```

```
# reading a file
# (4) readlines() approach
flanders_file = open('/content/drive/My Drive/Python - Dimensionless/Python Basics/flanders.txt', 'r')
line = flanders_file.readlines()
# returns a list of strings where each string is a line
print(line)
flanders_file.close()
```

```
numbers_list = []
flag = False
print('Enter a number to add to list, or \'done\' to exit.')
while True:
  user_input = input('>>>')
  if user_input.lower() == 'done':
    break
  else:
    try:
      numbers_list.append(float(user_input))
    except:
      print('You have entered other than a number or \'done\'.')
      print('exiting now...')
      flag = True
      break

if flag == True:
  pass
elif len(numbers_list) == 0:
  print('No number added to list.')    
else:
  print('Average of numbers is', round(sum(numbers_list)/len(numbers_list),2))
```
